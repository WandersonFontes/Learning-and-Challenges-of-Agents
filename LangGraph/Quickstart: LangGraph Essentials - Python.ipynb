```python
from typing_extensions import TypedDict
from langgraph.graph import StateGraph, START, END
```

```python
# Define the state schema for the graph
class State(TypedDict):
    value: int
```

```python
# Instantiate the StateGraph
graph_builder = StateGraph(State)
```

```python
# Define node functions (stubbed for brevity)
def node_a(state: State) -> State:
    return {"value": state["value"] + 1}

def node_b(state: State) -> State:
    return {"value": state["value"] * 2}

def condition_branch(state: State) -> str:
    return "node_b" if state["value"] + 2 == 0 else "node_a"
```

```python
# Add nodes to the graph
graph_builder.add_node("A", node_a)
graph_builder.add_node("B", node_b)

# Set the entry point of the graph
graph_builder.set_entry_point("A")

# Add conditional edges from node A
graph_builder.add_conditional_edges(
    "A",
    condition_branch,
    {"node_a": "A", "node_b": "B"} # Mapping returned value to next node
)

# Add a regular edge from node B to node A
graph_builder.add_edge("B", "A")

# Set the finish point of the graph
graph_builder.set_finish_point("B")
```

```python
# Compile the graph for execution
compiled_graph = graph_builder.compile()

# Example invocation (optional, for demonstration)
print(compiled_graph.invoke({"value": 1}))
```
