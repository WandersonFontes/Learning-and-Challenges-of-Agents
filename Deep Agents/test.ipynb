## Teoria

O que são agentes profundos?

Resposta: 

Agentes capazes de planejar, usar subagentes e aproveitar sistemas de arquivos para tarefas complexas.

Quando usar agentes profundos?

Resposta:

Use agentes profundos quando precisar de agentes que possam:

1 - Lidar com tarefas complexas e de várias etapas que exigem planejamento e decomposição.
2 - Gerencie grandes quantidades de contexto por meio de ferramentas do sistema de arquivos.
3 - Delegar tarefas a subagentes especializados para isolamento de contexto
4 - Manter a memória ativa em diferentes conversas e tópicos.

Para casos de uso mais simples, considere usar o LangChain create_agentou criar um fluxo de trabalho LangGraph personalizado .
